```python
board = { "Benny", "Hans", "Tine", "Mille", "Torben", "Troels", "Søren" }
management = { "Tine", "Trunte", "Rane" }
employees = { "Niels", "Anna", "Tine", "Ole", "Trunte", "Bent", "Rane", "Allan", "Stine", "Claus", "James", "Lars" }

board_not_employee = board.difference(employees)
print("Board not employees:", board_not_employee)

board_also_employee = board.intersection(employees)
print("Board also employees:", board_also_employee)

management_also_board = management.intersection(board)
print("Management also board:", management_also_board)

management_also_employee = management.intersection(employees)
print("Management also employees:", management_also_employee)

employee_also_management_and_board = employees.intersection(management, board)
print("Employees also management and board:", employee_also_management_and_board)

dictionary_data = {'a': 'Alpha', 'b': 'Beta', 'g': 'Gamma'}
list_of_tuples = list(dictionary_data.items())
print("List of tuples from dictionary:", list_of_tuples)

vowels_en = {'a', 'e', 'i', 'o', 'u', 'y'}
vowels_dk = {'a', 'e', 'i', 'o', 'u', 'y', 'æ' ,'ø', 'å'}
vowels_union = vowels_en.union(vowels_dk)
print("Vowels union:", vowels_union)

vowels_symmetric_difference = vowels_en.symmetric_difference(vowels_dk)
print("Vowels symmetric difference:", vowels_symmetric_difference)

vowels_difference = vowels_en.difference(vowels_dk)
print("Vowels difference:", vowels_difference)

vowels_disjoint = vowels_en.isdisjoint(vowels_dk)
print("Vowels disjoint:", vowels_disjoint)

vowels_union_operators = vowels_en | vowels_dk
print("Vowels Union (Operators):", vowels_union_operators)

vowels_symmetric_difference_operators = vowels_en ^ vowels_dk
print("Vowels Symmetric Difference (Operators):", vowels_symmetric_difference_operators)

vowels_difference_operators = vowels_en - vowels_dk
print("Vowels Difference (Operators):", vowels_difference_operators)

# der findes ingen operator der kan disjoint'e, man skal bruge metoden

months = {
    "JAN" : 1, "FEB" : 2, "MAR" : 3,
    "APR" : 4, "MAY" : 5, "JUN" : 6,
    "JUL" : 7, "AUG" : 8, "SEP" : 9,
    "OCT" : 10, "NOV" : 11, "DEC" : 12,
}

def get_date(date):
    dateParts = date.split("-")
    if len(dateParts) != 3:
        return None
    
    day = int(dateParts[0])
    month_abbrev = dateParts[1].upper()
    year = int(dateParts[2])

    if month_abbrev not in months:
        return None
    
    month = months[month_abbrev]

    if year < 100:
        if year < 50:
            year += 2000
        else:
            year += 1900
    
    return (year, month, day)

print("Date test 1:", get_date("03-SEP-89"))
print("Date test 2:", get_date("07-SEP-23"))

invited_friends = { "Niels", "Henning" }
rsvp_friends = { "Niels", "Børge" }

def invite_check(invited, rsvp):
    if isinstance(invited, set) and isinstance(rsvp, set):
        invited_no_rsvp = invited.difference(rsvp)
        print("Invited, no RSVP:", invited_no_rsvp)

        rsvp_no_invite = rsvp.difference(invited)
        print("RSVP, no invite:", rsvp_no_invite)

        invited_and_rsvp = invited.intersection(rsvp)
        print("Invited and RSVP:", invited_and_rsvp)
    else:
        return None

invite_check(invited_friends, rsvp_friends)
```
